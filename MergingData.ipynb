combines `series` with `geoseries` or `dataframe` with `geodataframe`  

# spatial join

observations from two `GeoSeries` or `GeoDataFrame`   
are combined based on their <i>spatial relationship</i> to one another.

```python
import geopandas
import geodatasets
chicago = geopandas.read_file(geodatasets.get_path("geoda.chicago_commpop"))
groceries = geopandas.read_file(geodatasets.get_path("geoda.groceries"))
print("\n Chicago \n")
print(chicago.info())
print("\n Groceries \n")
print(groceries.info())
#attribute join
chicago_shapes = chicago[['geometry','NID']]
chicago_names = chicago[['community','NID']]

print("\n chicago_shapes \n")
print(chicago_shapes.info())
print("\n chicago_names \n")
print(chicago_names.info())

#spatial join
chicago = chicago[["geometry",'community']].to_crs(groceries.crs)
print("\n chicagoo spatial join")
print(chicago.info())
```

# appeding

```python
import pandas as pd
#append series
joined = pd.concat([chicago.geometry,groceries.geometry])
print("\n joined \n")

print(joined.info())

#append dataframe

douglas = chicago.query("community =='DOUGLAS'")
print("\n douglas \n")

print(douglas)
oakland = chicago.query("community =='OAKLAND'")
print("\n oakland \n")

print(oakland)

douglas_oakland = pd.concat([douglas,oakland])

print("\n douglas_oakland\n")
print(douglas_oakland)
douglas_oakland.plot()
```

## This uses binary predicate

`predicate` attribute specifies whelther or not to join the attribute of one object to another based on their geometric relationship.
`Shapely` predicate keywords allowed:
    intersects

    contains

    within

    touches

    crosses

    overlaps
    
`how` indicates how the join will be like: `left`, `right`, `inner`. 

You can use `buffer()` and do `sjoin()` with specific `predicate`


```python
groceries_with_communities = groceries.sjoin(chicago, how="inner")
groceries_with_communities.head()
```

# Nearest join with `sjoin_nearest()`

## max_distance

The max_distance argument specifies a maximum search radius for matching geometries. This can have a considerable performance impact in some cases. If you can, it is highly recommended that you use this parameter.

## distance_col

If set, the resultant GeoDataFrame will include a <u>**column with this name containing the computed distances**</u> between an input geometry and the nearest geometry.
