### `dissolved` is used instead of `groupby`, a way to `groupby`

`dissolve()` can be thought of as doing three things:

    it dissolves all the geometries within a given group together into a single geometric feature (using the unary_union method), and

    it aggregates all the rows of data in a group using groupby.aggregate, and

    it combines those two results.


```python
import geodatasets
import geopandas

nepal = geopandas.read_file(geodatasets.get_path("geoda.nepal"))

nepal = nepal.rename(columns={"name_2":"zone"})

nepal[["zone","geometry"]].head()
```

### by default `dissolve()` with pass first to `groupy.aggregate()`

```python
nepal_zone = nepal[["zone","geometry"]]

zones = nepal_zone.dissolve(by="zone") 
zones.plot()
zones.head()
```

## `plot()` parameters of note:

### <u>column: str, np.array, pd.Series (default None)</u>

#### The name of the dataframe column, np.array, or pd.Series to be plotted. 
#### If np.array or pd.Series are used then it must have same length as dataframe. 
#### Values are used to color the plot. Ignored if color is also set.

    

### <u>scheme: str (default None)</u>

#### Name of a choropleth classification scheme (requires mapclassify). A mapclassify.MapClassifier object will be used under the hood. 
#### Arguments can be passed in classification_kwds.


```python
nepal_pop = nepal[["zone","geometry","population"]]
population = nepal_pop.dissolve(by="zone",aggfunc="sum")

population.plot(column = "population",scheme="quantiles", cmap="YlOrRd")
population.head()
```

`aggfunc` accepts  summary statistic options as allowed by pandas.groupby including:

    ‘first’

    ‘last’

    ‘min’

    ‘max’

    ‘sum’

    ‘mean’

    ‘median’

    function

    string function name

    list of functions and/or function names, e.g. [np.sum, ‘mean’]

    dict of axis labels -> functions, function names or list of such.


For example to get the number of distric on each zone,, as well as the populations of the largest and smallest district of each,  
you can aggregate the 'name' column using 'count', and the 'pop_est' column using 'min' and 'max':

```python
zones = nepal.dissolve(by ="zone",aggfunc={
"district" : "count",
"population" : ["min","max"]
})
zones.head()
```
